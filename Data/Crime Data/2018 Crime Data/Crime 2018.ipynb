```python
import pandas as pd
import numpy as np
import csv
import json
from pprint import pprint
```

```python
df_complaints = pd.read_csv("NYPD_Complaint Data_YTD_2018.csv")[['RPT_DT','ADDR_PCT_CD','BORO_NM', 'KY_CD','LAW_CAT_CD','OFNS_DESC','PD_DESC','Latitude','Longitude']]
df_complaints.head()
```

```python
df_complaints = df_complaints.dropna(axis='rows')
df_complaints.head()
```

```python
df_Crime2018 = df_complaints[(df_complaints['RPT_DT']>='01/01/2018')& (df_complaints['RPT_DT']<='12/31/2018')]
df_Crime2018.head()
```

```python
df_Crime2018['Lat_Lng'] = df_Crime2018[['Latitude','Longitude']].values.tolist()
df_Crime2018 = df_Crime2018.drop(['Latitude', 'Longitude'], axis=1)
df_Crime2018.head()
```

```python
df_Crime2018.to_csv("2018_CrimeData.csv", sep=',')
```

```python
df_GrandLarceny2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('LARCENY')]
df_GrandLarceny2018.to_csv("GrandLarceny2018.csv", sep=',')

df_Assault2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('ASSAULT')]
df_Assault2018.to_csv("Assault2018.csv", sep=',')

df_Vehicles2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('VEHICLE')]
df_Vehicles2018.to_csv("Vehicles2018.csv", sep=',')

df_Driving2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('DRIVING')]
df_Driving2018.to_csv("Driving2018.csv", sep=',')

df_Drugs2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('DRUGS|DRUG')]
df_Drugs2018.to_csv("Drugs2018.csv", sep=',')

df_Burglary2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('BURGLARY')]
df_Burglary2018.to_csv("Burglary2018.csv", sep=',')

df_SexCrimes2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('RAPE|SEX')]
df_SexCrimes2018.to_csv("SexCrimes2018.csv", sep=',')

df_Trespass2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('TRESPASS')]
df_Trespass2018.to_csv("Trespass2018.csv", sep=',')

df_Kidnapping2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('KIDNAPPING')]
df_Kidnapping2018.to_csv("Kidnapping2018.csv", sep=',')

df_GambProstHom2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('GAMBLING|PROSTITUTION|HOMICIDE')]
df_GambProstHom2018.to_csv("GambProstHom2018.csv", sep=',')

df_Other2018 = df_Crime2018[df_Crime2018['OFNS_DESC'].str.contains('OFFENSES|SAFETY')]
df_Other2018 .to_csv("Other2018.csv", sep=',')
```
