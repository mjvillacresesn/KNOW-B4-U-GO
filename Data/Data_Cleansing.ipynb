```python
# Dependencies 
import pandas as pd
```

```python
# import/read csv files
jan_csv = pd.read_csv("Data/listings_01.csv", encoding="ISO-8859-1")
feb_csv = pd.read_csv("Data/listings_02.csv", encoding="ISO-8859-1")
mar_csv = pd.read_csv("Data/listings_03.csv", encoding="ISO-8859-1")
apr_csv = pd.read_csv("Data/listings_04.csv", encoding="ISO-8859-1")
may_csv = pd.read_csv("Data/listings_05.csv", encoding="ISO-8859-1")
jun_csv = pd.read_csv("Data/listings_06.csv", encoding="ISO-8859-1")
jul_csv = pd.read_csv("Data/listings_07.csv", encoding="ISO-8859-1")
aug_csv = pd.read_csv("Data/listings_08.csv", encoding="ISO-8859-1")
sep_csv = pd.read_csv("Data/listings_09.csv", encoding="ISO-8859-1")
oct_csv = pd.read_csv("Data/listings_10.csv", encoding="ISO-8859-1")
nov_csv = pd.read_csv("Data/listings_11.csv", encoding="ISO-8859-1")
dec_csv = pd.read_csv("Data/listings_12.csv", encoding="ISO-8859-1")

columns = [ "id",
            "neighbourhood_cleansed",
            "zipcode",
            "latitude",
            "longitude",
            "accommodates"
            "price",
            "review_scores_rating"  ]

csv_list = [jan_csv, feb_csv, mar_csv, apr_csv, may_csv, jun_csv, jul_csv, aug_csv, sep_csv, oct_csv, nov_csv, dec_csv]
filtered_list = []

for month in csv_list:
    month = month[columns]
    
    #add to filtered_list
    filtered_list.append(month)
```

```python
# create empty dataframe with same columns
data_2018 = pd.DataFrame(columns=columns)

# append csv files based on same columns
for month in filtered_list:
    data_2018 = data_2018.append(month, ignore_index=True)
    
data_2018.head(20)
```

```python
# drop rows with NaN values and values in zipcode column that do not match criteria
data_2018 = data_2018.dropna(axis=0, how='any')
data_2018 = data_2018.drop(data_2018[data_2018['zipcode'].astype(str).map(len) != 5].index)

data_2018.head(20)
```

```python
# check types of each metric
data_2018.dtypes
```

```python
# get rid of ($) from price column
data_2018["price"] = data_2018["price"].replace('[\$,]','', regex=True).astype(float)

# convert metrics to numeric
numeric_col = [ "zipcode",
                "price" ]

for col in numeric_col:
    data_2018[col] = pd.to_numeric(data_2018[col])

data_2018.dtypes
```

```python
# clean up column names
data_2018 = data_2018.rename(columns={
    'neighbourhood_cleansed':'neighborhood',
    'review_scores_rating':'rating'
})
```

```python
# export data to csv file
data_2018.to_csv('Data/data_2018.csv')
```
